VISUAL PRESENTATION



```python
##colors = ['green', 'red', 'yellow', 'black', 'pink'] 
plt.figure(figsize= (8,5))
plt.bar(campaign_df['CAMPAIGNS'], campaign_df['ACCEPTANCES'], color= ['green', 'red', 'yellow', 'black', 'pink'] )
plt.title('BEST PERFORMING CAMPAIGN')
plt.ylabel('NUMBER OF ACCEPTANCES')
plt.xlabel('CAMPAIGNS')
plt.show()
```

```python
## ARE HIGHER SPENDING CUSTOMERS MORE RESPONSIVE?
campaign_cols = ["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5"]

df["Responsive"] = df[campaign_cols].sum(axis=1) > 0

spend_cols = [
    "MntWines","MntFruits","MntMeatProducts", "MntFishProducts","MntSweetProducts","MntGoldProds"]

df["TotalSpend"] = df[spend_cols].sum(axis=1)


```

```python
#spend_cols = [
    #"MntWines","MntFruits","MntMeatProducts", "MntFishProducts","MntSweetProducts","MntGoldProds"]

#df["TotalSpend"] = df[spend_cols].sum(axis=1)
#df.groupby("Responsive")["TotalSpend"].mean()
df.groupby("Responsive")["TotalSpend"].mean()
```

```python
sns.barplot(
    x="Responsive",
    y="TotalSpend",
    data=df,
    palette =['black','yellow'])

plt.title("Average Spending by Campaign Responsiveness")
plt.xlabel("Responsive to Campaign")
plt.ylabel("Average Total Spend")
plt.show()
```
